## ‚ú® Introdu√ß√£o

O presente estudo tem como objetivo avaliar e explorar a rela√ß√£o entre os dados de desmatamento no estado do Par√° e os indicadores socioecon√¥micos dos seus munic√≠pios. A atividade foi desenvolvida para o processo seletivo do projeto ZettaLab, na trilha de Governan√ßa de Dados.

## üîç Fontes de Dados

- **Desmatamento (INPE - TerraBrasilis):** dados de 2007 a 2024 por munic√≠pio  
- **Socioecon√¥micos (IBGE):** IDHM, PIB per capita, popula√ß√£o, escolariza√ß√£o (2010 a 2024)  
- As bases foram integradas por c√≥digo IBGE

## üõ†Ô∏è Metodologia

1. Corre√ß√£o dos dados decimais (INPE)  
2. Convers√£o de colunas num√©ricas  
3. Merge por c√≥digo IBGE  
4. Cria√ß√£o do indicador: √°rea desmatada per capita  
5. Remo√ß√£o de 2007 por padr√£o (ac√∫mulo de dados)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar dados integrados
df = pd.read_csv("saida/base_completa_merge.csv")
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['areakm'] = pd.to_numeric(df['areakm'], errors='coerce')
df['idhm_2010'] = pd.to_numeric(df['idhm_2010'], errors='coerce')
df['idhm_cat'] = df['idhm_cat'].astype(str)
```

```python
variaveis = ['areakm', 'populacao_2022', 'pib_per_capita_2021', 'idhm_2010', 'desmatamento_per_capita']
correlacao = df[variaveis].corr()

plt.figure(figsize=(7, 5))
sns.heatmap(correlacao, annot=True, cmap='coolwarm')
plt.title("Correla√ß√£o entre Indicadores Socioecon√¥micos e Desmatamento")
plt.show()
```

## üìà Evolu√ß√£o do Desmatamento (2008‚Äì2024)

Abaixo, visualizamos a √°rea total desmatada por ano (2007 exclu√≠do por padr√£o):


```python
df_filtrado = df[df['year'] > 2007]
evolucao = df_filtrado.groupby('year')['areakm'].sum().reset_index()

plt.figure(figsize=(10, 4))
sns.lineplot(data=evolucao, x='year', y='areakm', marker='o')
plt.title("Evolu√ß√£o da √Årea Total Desmatada no Par√°")
plt.xlabel("Ano")
plt.ylabel("√Årea Desmatada (km¬≤)")
plt.grid(True)
plt.show()
```

## üìä Desmatamento Total por Categoria de IDHM

```python
# Remover valores ausentes ou inv√°lidos em idhm_cat
df_barras = df_filtrado[df_filtrado['idhm_cat'].isin(['Baixo', 'M√©dio', 'Alto'])]
df_barras = df_barras.groupby('idhm_cat')['areakm'].sum().reset_index()

plt.figure(figsize=(6,4))
sns.barplot(data=df_barras, x='idhm_cat', y='areakm', hue='idhm_cat', palette='Set2', legend=False)
plt.title("Desmatamento Total por Categoria de IDHM")
plt.xlabel("Categoria de IDHM")
plt.ylabel("√Årea Desmatada (km¬≤)")
plt.show()
```

## ‚úÖ Conclus√µes

- Os munic√≠pios com IDHM mais alto concentram maior √°rea desmatada total.  
- N√£o h√° uma rela√ß√£o linear clara entre IDHM e desmatamento, mas sim contextual.  
- Os dados fixos do IBGE funcionam bem como perfil base de compara√ß√£o.  
- O ano de 2007 foi desconsiderado para evitar distor√ß√µes (dados acumulados).
